```python
# üìä Load the Synthetic Transaction Dataset
import pandas as pd

# Load the CSV you generated earlier
df = pd.read_csv("../data/synthetic_finory_transactions.csv")

# Display first 5 rows
df.head()
```

```python
# üîç Explore structure and summary
print("‚úÖ Shape of dataset:", df.shape)
print("\nüß† Column data types:")
print(df.dtypes)

print("\nüìà Summary statistics:")
print(df.describe(include='all'))
```

```python
# üßº Cleanliness check
print("‚ùì Missing values:")
print(df.isnull().sum())

print("\nüîÅ Duplicates:", df.duplicated().sum())
```

```python
# üìä Explore key categorical distributions
print("üõçÔ∏è Top Vendors:")
print(df["vendor"].value_counts().head(10))

print("\nüìÇ Category Distribution:")
print(df["category"].value_counts())

print("\nüí≥ Payment Methods:")
print(df["payment_method"].value_counts())
```

```python
# üîç Category Distribution
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 20))
df['category'].value_counts().sort_values().plot(kind='barh', color='skyblue')
plt.title('Transaction Count by Category')
plt.xlabel('Count')
plt.ylabel('Category')
plt.tight_layout()
plt.show()
```

```python
# üíµ Distribution of Transaction Amounts
plt.figure(figsize=(10, 5))
df['amount'].hist(bins=50, color='lightgreen')
plt.title('Distribution of Transaction Amounts')
plt.xlabel('Amount ($)')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# üì¶ Boxplot of Amounts by Category
import seaborn as sns

# Optional: sort categories by median amount
sorted_categories = (
    df.groupby("category")["amount"]
    .median()
    .sort_values(ascending=False)
    .index
)

plt.figure(figsize=(10, 20))  # Wider height for scrolling
sns.boxplot(data=df, y="category", x="amount", order=sorted_categories)
plt.title("Transaction Amounts by Category (Horizontal)")
plt.xlabel("Amount ($)")
plt.ylabel("Category")
plt.grid(axis="x")
plt.tight_layout()
plt.show()
```

```python
# üí≥ Average Spend by Payment Method
avg_by_method = df.groupby('payment_method')['amount'].mean().sort_values()

plt.figure(figsize=(10, 5))
avg_by_method.plot(kind='barh', color='coral')
plt.title('Average Transaction Amount by Payment Method')
plt.xlabel('Average Amount ($)')
plt.ylabel('Payment Method')
plt.grid(axis='x')
plt.tight_layout()
plt.show()
```
